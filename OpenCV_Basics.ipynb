#### Addition

```python
import numpy as np
# Determine the new size (use the smaller of the two images to avoid cutting off parts)
new_height = min(img1.shape[0], img2.shape[0])
new_width = min(img1.shape[1], img2.shape[1])

# Resize the images to the new size
img1_resized = cv2.resize(img1, (new_width, new_height))
img2_resized = cv2.resize(img2, (new_width, new_height))

# Add the images using NumPy (clipping the values to be in the valid range for image data)
result = img1_resized + img2_resized
```

# Blurring of Image

```python

# Load the image from the specified path
img1 = cv2.imread(r"D:\work\NullClass\Age and Gender Detector\Resources\Data\brain.jpg")

# Convert the image from BGR (default in OpenCV) to HSV color space
img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

# Set the size of the kernel for the blur effect
figure_size = 7

# Apply a blur effect to the image using the specified kernel size
result = cv2.blur(img, (figure_size, figure_size))

# Create a figure with a specified size for displaying the images
plt.figure(figsize=(11, 6))

# Display the original image (converted back to RGB for correct color representation in matplotlib)
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
plt.title("Original")
plt.xticks([]), plt.yticks([])  # Hide the x and y axis ticks

# Display the blurred image (converted back to RGB for correct color representation in matplotlib)
plt.subplot(122)
plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB))
plt.title("Blurred")
plt.xticks([]), plt.yticks([])  # Hide the x and y axis ticks

# Show the plot with the original and blurred images
plt.show()
```

# Zooming of Image

```python
img1 = cv2.imread(r"D:\work\NullClass\Age and Gender Detector\Resources\Data\lena.png")
```

```python
# Get the height and width of the original image
(height, width) = img1.shape[0:2]

# Resize the image to 1.5 times its original size using cubic interpolation
zoom = cv2.resize(img1, None, fx=1.5, fy=1.5, interpolation=cv2.INTER_CUBIC)

# Display the resized image
plt.imshow(zoom)
plt.title("Zoomed Image")
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```

```python
zoom= cv2.resize(img1,(int(width*2),int(height*2)),interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

# Rotating the Image

```python
# Create a rotation matrix for rotating the image around its center
# Parameters: center point, angle of rotation, scale factor
m = cv2.getRotationMatrix2D((width / 2, height / 2), 30, 1)

# Apply the affine transformation (rotation) to the image
# Parameters: source image, rotation matrix, size of the output image
rotate = cv2.warpAffine(img1, m, (width, height))

# Display the rotated image
plt.imshow(rotate)  
plt.title("Rotated Image")
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```

# Translating an Image

```python
# Define the translation matrix
# Parameters: [1, 0, tx], [0, 1, ty]
# This matrix translates the image by tx pixels along the x-axis and ty pixels along the y-axis
m = np.float32([[1, 0, 50], [0, 1, 100]])

# Apply the affine transformation (translation) to the image
# Parameters: source image, translation matrix, size of the output image
dst = cv2.warpAffine(img1, m, (width, height))

# Display the translated image
plt.imshow(dst)
plt.title("Translated Image")
plt.axis('off')  # Hide the axes for better visualization
plt.show()
```
